## Check Nulls

```python
nul = pd.DataFrame(dataset.isnull().sum(), columns=["Nulls Count"])

nul["Nulls Percentage"] = list(map(lambda x : f"{x:.2f}%", (dataset.isnull().sum() / dataset.shape[0])*100))

nul.sort_values(by="Nulls Percentage", ascending=False)
```
